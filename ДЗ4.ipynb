Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
# Готовим данные для графика
# Найдем зависимость цены от кол-ва спален

data31 = df[['price', 'bedrooms']]
data31 = data31.groupby('bedrooms').mean().reset_index()
data31
```

```python
plt.figure(figsize=(6, 4))

plt.plot(data31['bedrooms'], data31['price'])

plt.title('Зависимость цены от кол-ва спален')
plt.xlabel('Спален')
plt.ylabel('Цена');

```

Цена максимальна в диапазоне от 7 до 10 спален

```python
# Найдем зависимость цены от кол-ва этажей

data32 = df[['price', 'floors']]
data32 = data32.groupby('floors').mean().reset_index()

data32
```

```python
plt.figure(figsize=(6, 4))

plt.bar(data32['floors'], data32['price'])

plt.title('Цена от кол-ва этажей')
plt.xlabel('Кол-во этажей')
plt.ylabel('Цена');
```

Цена максимальна при кол-ве этажей = 2.5

```python
# Найдем зависимость цены от вида на воду

data33 = df[['price', 'waterfront']]
data33 = data33.groupby('waterfront').mean().reset_index()

data33
```

```python
# Строим диаграмму из столбцов

plt.bar(data33['waterfront'], data33['price'])
plt.title('Цена от вида на воду')
plt.xlabel('Вид  на воду')
plt.ylabel('Цена');
```

Вид на воду существенно увеличивает цену
